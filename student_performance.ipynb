```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student-mat.csv', sep=';')

df.head()
```

```python
print("Shape of dataset:", df.shape)
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

```python
average_grade = df['G3'].mean()

print("Average Final Grade (G3):", average_grade)
```

```python
above_15 = df[df['G3'] > 15]

print("Number of students scoring above 15:", len(above_15))
```

```python
correlation = df['studytime'].corr(df['G3'])

print("Correlation between Study Time and G3:", correlation)
```

```python
gender_avg = df.groupby('sex')['G3'].mean()

print("Female Average G3:", gender_avg['F'])
print("Male Average G3:", gender_avg['M'])
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
gender_avg.plot(kind='bar')

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade (G3)")
plt.show()
```
